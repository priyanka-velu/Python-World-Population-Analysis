```python
# get high level info on each column

df.info()

# we have 234 columns until 2022 population where we start losing some rows values
# we know each columns' data types
```

```python
# get statistical info on each column

df.describe()

# we observe the count, mean, std, min, max, and quartiles
# there is a country with a population of 510
# there is a country with a max population of 1.42 billion
```

```python
# we need to see how many misssing values that could skew the dataset

df.isnull().sum()

# .isnull().sum() gives us missing data in each column
# should we populatate these with a median value? delete countries? let's start thinking of these questions
# need to find outliers and missing values during EDA
```

```python
# find the unique values in each of these columns

df.nunique()


# double check that there are 7 continents
# each rank should be unique
# all large populations should also be different numbers
# the world population percentage is much lower cause its a percentage a lot of them are really low (ie.0.01, 0.02)
```

```python
# countries with the largest 2022 population

df.sort_values(by='2022 Population', ascending=False).head(10)

# .sort_values(by="2022 population').head : sorts by the lowest population first
# we can change to ascending to get the highest populations
# the top 5 is china, india, us, indonesia, pakistan
```

```python
# countries with the largest World Population Percentage

df.sort_values(by='World Population Percentage', ascending=False).head(10)

# china is composed of 17.88% of the world's population
# india is composed of 17.77% of the world's population
# US is composed of 4.24% of the world's population
```

```python
# .corr: compare every column with every other column and seeing how closely correlated they are
# only select data types with a number format

df_numeric = df.select_dtypes(include=[np.number])
df_numeric.corr()

# we observe that all population will be exactly correlated because they are steadily increasing
# we observe that the area is somewhat correlated because a country may have a large population and a small area or vice versa
```

```python
# visualize correlation using a heatmap

sns.heatmap(df_numeric.corr(), annot=True)
plt.rcParams['figure.figsize'] = (20,7)
plt.show()

# population is highly correlated except density, area, and growth rate (rows) with population numbers
# we thought area would be more correlated cause more population means more area but not the case so it means that either population is increasing vastly or decreasign vastly compared to the area
```

```python
# are there certain continents that have grown faster than others?
# let's plot a graph that shows the population size growth for each continent

# only select float and int from continent

numeric_df = df.select_dtypes(include=[float, int])
numeric_df['Continent'] = df['Continent']

# groupby function for numeric values only
# get mean population values grouped by continent

mean_values = numeric_df.groupby('Continent').mean().sort_values(by='2022 Population', ascending = False)
mean_values

# we observe that asia has the highest population with oceania on the bottom, denstiy in asia is much higher
# we observe that oceania has a substantial population
```

```python
# search for different countries in continent oceania

df[df['Continent'].str.contains('Oceania')]
```

```python
# get columns

df.columns
```

```python
# create a new dataframe to plot the mean population of each continent, sorted from highest to lowest

df2 = numeric_df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by='2022 Population', ascending = False)
df2
```

```python
# which continent has grown the fastest?
# visualize population growth in each country

df2.plot()

# we need to transpose so we can look at the population growth of each continent
```

```python
# .transpose()

df3 = df2.transpose()
df3
```

```python
df3.plot(figsize=(12,6))

# population in asia exploded in 2000
# other continents stayed stagnant
# high level overview of population growth over time
```

```python
# create a boxplots to visualize outliers
df.boxplot(figsize=(20,10))

# outliers are expected when most countries are smaller
# 2022 population outliers most likely exist as the population of china, india, US, etc
```
